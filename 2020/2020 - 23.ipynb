```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    
    def __str__(self):
        return(str(self.data))

class CreateList:
    def __init__(self):
        self.head = Node(None)
        self.tail = Node(None)
        self.head.next = self.tail
        self.tail.next = self.head
        self.max_data = 0
        self.node_list = {}
        
    def add(self, data):
        newNode = Node(data)
        self.max_data = max(data, self.max_data)
        if self.head.data is None:
            self.head = newNode
            self.tail = newNode
            newNode.next = self.head
        else:
            self.tail.next = newNode
            self.tail = newNode
            self.tail.next = self.head
        self.node_list[data] = newNode
    
    def display(self, start_num=None):
        if start_num == None:
            current = start_node = self.head
        else:
            start_node = current = self.node_list[start_num]
        if self.head is None:
            print('List is empty')
            return
        else:
            print(current.data, end=' ')
            while(current.next != start_node):
                current = current.next
                print(current.data, end=' ')
            print()
    
    def one_turn(self, current_node):
        cut1 = current_node.next
        cut2 = cut1.next
        cut3 = cut2.next
        cut_list = [cut1.data, cut2.data, cut3.data]
        destination = current_node.data - 1
        while destination <= 0 or destination in cut_list:
            if destination <= 0:
                destination = self.max_data
            else:
                destination -= 1
        current_node.next = cut3.next
        cut3.next = self.node_list[destination].next
        self.node_list[destination].next = cut1
        return(current_node.next)
    
    def play_cups(self, curr_node, moves):
        for i in range(moves):
            curr_node = self.one_turn(curr_node)

    def part1_answer(self):
        current = start_node = self.node_list[1]
        result = ''
        while current.next != self.node_list[1]:
            current = current.next
            result += str(current.data)
        return(result)

    def part2_answer(self):
        num1 = self.node_list[1].next.data
        num2 = self.node_list[1].next.next.data
        return(num1 * num2)
            
def build_lists(inp):
    inp = [int(c) for c in inp]
    cl = CreateList()
    for c in inp:
        cl.add(c)
    return(cl)

actual_input = '156794823'

inp = list(actual_input)
cl = build_lists(inp)
cl.play_cups(cl.head, 100)
print('Part 1 Answer:',cl.part1_answer())

big_input = [int(c) for c in actual_input]
big_input = big_input + list(range(max(big_input)+1, 1000001))
big_cl = build_lists(big_input)
big_cl.play_cups(big_cl.head, 10000000)
print('Part 2 Answer:', big_cl.part2_answer())
```
