```python
def create_decks(filename):
    with open(filename, 'r') as f:
        data = f.read().splitlines()
    middle = data.index('')
    deck1 = [int(c) for c in data[1:middle]]
    deck2 = [int(c) for c in data[middle+2:]]
    return(deck1, deck2)

def score_deck(d):
    return(sum([(len(d) - i) * card for i, card in enumerate(d)]))

def combat_turn(deck1, deck2, turn=0):
    player1_card = deck1.pop(0)
    player2_card = deck2.pop(0)
    if player1_card > player2_card:
        deck1.append(player1_card)
        deck1.append(player2_card)
    else:
        deck2.append(player2_card)
        deck2.append(player1_card)

def combat_game(deck1, deck2):
    turn = 0
    while len(deck1) > 0 and len(deck2) > 0:
        turn += 1
        combat_turn(deck1, deck2, turn)
    if len(deck1) > 0:
        return(1, score_deck(deck1), turn)
    else:
        return(2, score_deck(deck2), turn)

def recursive_combat_turn(deck1, deck2, previous_rounds = []):
    if (deck1, deck2) in previous_rounds:
        return(1)
    else:
        player1_card = deck1.pop(0)
        player2_card = deck2.pop(0)
        if len(deck1) >= player1_card and len(deck2) >= player2_card:
            sub_deck1 = deck1[:player1_card]
            sub_deck2 = deck2[:player2_card]
            winner, _, _ = recursive_combat_game(sub_deck1, sub_deck2)
            if winner == 1:
                deck1.append(player1_card)
                deck1.append(player2_card)
            else:
                deck2.append(player2_card)
                deck2.append(player1_card)
            return(0)
        else:
            if player1_card > player2_card:
                deck1.append(player1_card)
                deck1.append(player2_card)
            else:
                deck2.append(player2_card)
                deck2.append(player1_card)
            return(0) # game continues
                
def recursive_combat_game(deck1, deck2):
    previous_rounds = []
    turn = 0
    game_over = False
    while not game_over:
        turn += 1
        turn_start_deck1 = deck1.copy()
        turn_start_deck2 = deck2.copy()
        turn_result = recursive_combat_turn(deck1, deck2, previous_rounds)
        previous_rounds.append((turn_start_deck1, turn_start_deck2))
        if turn_result == 1 or len(deck2) == 0:
            return(1, score_deck(deck1), turn)
        if len(deck1) == 0:
            return(2, score_deck(deck2), turn)

def solve_problem(filename):
    deck1, deck2 = create_decks(filename)
    winner, score, turns = combat_game(deck1.copy(), deck2.copy())
    print('Part 1 Answer: Player', winner, 'won with a score of', score, 'in', turns,'turns!')
      
    winner, score, turns = recursive_combat_game(deck1.copy(), deck2.copy())
    print('Part 2 Answer: Player', winner, 'won with a score of', score, 'in', turns,'turns!')

solve_problem('input_day22.txt')
```
